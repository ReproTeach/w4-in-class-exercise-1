### Line Plot

```python
# write your code here, you can add more cells if needed

import pandas as pd
import matplotlib.pyplot as plt

x = range(0,34)
y = df["python"]

plt.plot(x, y, color='green', linestyle='-', linewidth=1)
plt.title("Comm160DS Python Knowledge")
plt.xlabel("Answers")
plt.ylabel("Knowledge of Python Programming")
plt.show()
```

### Scatter Plot 

```python
# write your code here, you can add more cells if needed
import pandas as pd
import matplotlib.pyplot as plt

x = range(0,34)
y = df["datascience"]

plt.scatter(x, y, color='green')
plt.title("Comm160DS Knowledge of Datascience")
plt.xlabel("Students")
plt.ylabel("Knowledge of Datascience")
plt.xlim(0, 35)
plt.ylim(0, 4)
plt.yticks([0, 1, 2, 3, 4])
plt.show()
```

###  Bar Plot

```python
df["nlp"].value_counts()
```

```python
x = ['No', 'Yes']
y = df["nlp"].value_counts()

plt.bar(x, y, color=("green","blue"))

plt.title('Experience of NLP') 
plt.xlabel('Experience') 
plt.ylabel('Number of Students') 
plt.yticks([0, 2, 10, 20, 30, 32])

plt.show()
```
